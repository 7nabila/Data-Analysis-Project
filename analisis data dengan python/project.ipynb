```python
# Membuat dataframe baru dari data day
day_select = day_df[['season', 'yr', 'mnth', 'cnt', 'holiday', 'weekday']].copy()

# Membuat dataframe baru dari data hour
hour_select = hour_df[['season', 'yr', 'mnth', 'cnt', 'holiday', 'weekday']].copy()

# Menampilkan beberapa baris pertama dari dataframe baru day
print("Data 'day' yang dipilih:")
print(day_select.head())

# Menampilkan beberapa baris pertama dari dataframe baru hour
print("\nData 'hour' yang dipilih:")
print(hour_select.head())
```

<h3>EDA

```python
# 1. Statistik Deskriptif
print("Statistik Deskriptif untuk Data 'hour' yang Dipilih:")
print(hour_select.describe())

# 2. Distribusi Data
hour_select.hist(figsize=(10, 8), bins=20)
plt.suptitle("Distribusi Data untuk Data 'hour' yang Dipilih")
plt.show()

# 3. Korelasi
correlation_matrix_hour = hour_select.corr()
sns.heatmap(correlation_matrix_hour, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriks Korelasi untuk Data 'hour' yang Dipilih")
plt.show()
```

<h3> Visualization

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='season', y='cnt', data=day_select)
plt.title('Jumlah Penyewa Sepeda Berdasarkan Musim')
plt.xlabel('Musim')
plt.ylabel('Jumlah Penyewa Sepeda')
plt.show()
```

Musim dengan peminjaman sepeda paling banyak terjadi pada musim ke-3 (gugur)

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='weekday', y='cnt', hue='holiday', data=day_select)
plt.title('Perbandingan Penyewaan Sepeda di Hari Kerja dan Hari Libur')
plt.xlabel('Hari dalam Seminggu')
plt.ylabel('Jumlah Penyewa Sepeda')
plt.legend(title='Hari Libur', labels=['Tidak', 'Ya'])
plt.show()
```

jumlah penyewa pada hari kerja lebih stabil daripada hari libur

<h3> Conclusion:

Dari data yang sudah diolah diatas, peminjaman paling banyak terjadi pada musim gugur tertama pada hari kerja
